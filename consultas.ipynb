```python
# las 5 mayores
df = pd.read_sql("SELECT TOP (5) venta_id, categoria, valor "
                 "FROM ventas ORDER BY valor DESC", con)
display(df)
 
# las 5 siguientes (pagina dos)
df2 = pd.read_sql("SELECT venta_id, categoria, valor FROM ventas "
                  "ORDER BY valor DESC "
                                    "OFFSET 5 ROWS FETCH NEXT 5 ROWS ONLY", con)
display(df2)
```
